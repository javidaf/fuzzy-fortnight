```python
from tek5020_p2.utils import *
from tek5020_p2.__main__ import * 
train_image_path = r"..\tests\data\Bilde2.png"
test_image_path = r"..\tests\data\Bilde3.png"
```

## Step 

```python
size = 350
r1 = (950, 1200)
r2 = (2500, 800)
r3 = (400, 400)

regions = [
    Region(r1[0] - size, r1[1] - size, r1[0] + size, r1[1] + size, 1),
    Region(r2[0] - size, r2[1] - size, r2[0] + size, r2[1] + size, 2),
    Region(r3[0] - size, r3[1] - size, r3[0] + size, r3[1] + size, 3),
]
```

```python
normalize = True
train_image = load_image(train_image_path)
test_image = load_image(test_image_path)
X_train, y_train = get_training_data(train_image, regions,normalize=normalize)

classifier = MinimumErrorRateClassifier()
classifier.fit(X_train, y_train)
train_result = segment_image(train_image, classifier,normalize=normalize)
test_result = segment_image(test_image, classifier,normalize=normalize)
```

```python
train_visualization = visualize_segmentation(train_result, len(regions))
test_visualization = visualize_segmentation(test_result, len(regions))
train_with_regions = draw_regions(train_image, regions)
plt.imshow(train_with_regions)
plt.title("Training Image with Regions")
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(221)
train_with_regions = draw_regions(train_image, regions)
plt.imshow(train_with_regions)
plt.title("Training Image with Regions")

plt.subplot(222)
plt.imshow(train_visualization)
plt.title("Segmented Training Image")

plt.subplot(223)
plt.imshow(test_image)
plt.title("Test Image (Bilde3)")

plt.subplot(224)
plt.imshow(test_visualization)
plt.title("Segmented Test Image")

plt.tight_layout()
plt.show()
```
